```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Question 1

```python
Telco=pd.read_csv("Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv")
```

```python
Telco
```

```python
plt.figure(figsize=(8,8))
sns.countplot(Telco["gender"],hue=Telco["Churn"])
plt.title("Churn count vs Gender",fontsize=18)
```

# Question 2

```python
SCF=Telco.loc[(Telco["gender"]=="Female") & (Telco["SeniorCitizen"]==1)]
```

```python
SCF
```

```python
sns.countplot(SCF["gender"])
plt.title("Count of Senior Citizens")
```

# Question 3

```python
plt.figure(figsize=(8,8))
plt.scatter(Telco["tenure"],Telco["TotalCharges"],s=10,marker="*")
plt.xlabel("Tenure",fontsize=14)
plt.ylabel("Total Charges",fontsize=14)
plt.title("Relation between Tenure and Total charges",fontsize=18)
```
