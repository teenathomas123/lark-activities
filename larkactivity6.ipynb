# Question 2

```python

plt.figure(figsize=(10,6))
sns.barplot(x=data['Region'],y=data['GDP per capita (current US$)'],data=data)
plt.xticks(rotation=90)
plt.title("GDP per capita vs Region")
plt.show()
```

```python
sns.jointplot(x=data["GDP per capita (current US$)"], y= data["Region"],data=data)
plt.show()
```

# QUESTION 4

```python
data1=data.groupby("Region")[["GDP per capita (current US$)"]].mean()
```

```python
data1.sort_values('GDP per capita (current US$)')
```

```python
data1.nlargest(1,"GDP per capita (current US$)")
```

```python
data2=data.groupby("Region")[["International trade: Imports (million US$)"]].mean()
```

```python
data2.sort_values("International trade: Imports (million US$)").tail()
```

```python
data2.nlargest(1,"International trade: Imports (million US$)")
```
